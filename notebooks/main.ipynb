```python
import os

# 1. 파일 경로 설정
image_path = "score_images/315.jpg"

# 2. oemer가 시스템에 설치되어 있는지 확인하고 바로 실행
if os.path.exists(image_path):
    print(f"✅ 파일을 찾았습니다: {image_path}")
    print("🚀 OMR 분석을 시작합니다. 이 작업은 1~2분 정도 소요될 수 있습니다...")
    
    # 시스템 명령어로 oemer 실행
    !oemer "{image_path}"
    
    # 3. 결과 파일이 생성되었는지 확인
    # oemer는 보통 실행한 이미지와 같은 위치 혹은 현재 디렉토리에 파일을 만듭니다.
    print("\n📂 생성된 결과물 확인:")
    output_files = [f for f in os.listdir('.') if f.endswith(('.musicxml', '.mid'))]
    if not output_files:
        # 이미지 폴더 내부도 확인
        result_dir = os.path.dirname(image_path)
        output_files = [f for f in os.listdir(result_dir) if f.endswith(('.musicxml', '.mid'))]
    
    print(output_files if output_files else "⚠️ 파일이 아직 생성되지 않았습니다. 위 로그의 에러 메시지를 확인해 주세요.")
else:
    print(f"❌ 파일을 찾을 수 없습니다: {image_path}")
```

```python
import music21

# 1. (가정) 성공적으로 변환된 찬송가 XML 로드
# 지금은 파일이 없으니 music21 제공 샘플이나 빈 객체로 구조만 잡습니다.
def analyze_satb(xml_path):
    score = music21.converter.parse(xml_path)
    
    # 2. 성부 분리 (보통 4성부 악보는 Part가 나뉘어 있거나 Voice로 구분됨)
    parts = score.getElementsByClass(music21.stream.Part)
    
    voices = ["Soprano", "Alto", "Tenor", "Bass"]
    for i, part in enumerate(parts):
        print(f"--- {voices[i]} 파트 분석 ---")
        notes = part.flat.notes
        for n in notes[:5]: # 상위 5개 음표 확인
            if n.isNote:
                print(f"음높이: {n.pitch.nameWithOctave}, 길이: {n.duration.quarterLength}")
            elif n.isChord:
                print(f"화음 발견: {[p.nameWithOctave for p in n.pitches]}")

# 이 로직이 완성되면 각 성부의 MIDI 데이터를 AI 가창 모델(DiffSinger 등)에 
# 4번 던져서 4개의 목소리 파일을 만들고, 이를 하나로 합치면 'AI 찬양대'가 됩니다.
```
